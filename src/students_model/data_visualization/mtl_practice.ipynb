```python
import pandas as pd
import matplotlib.pyplot as plt
from student_analyzer import StudentAnalyzer

# Load data
df = pd.read_csv('../resources/cleaned_students.csv')

# Create analyzer instance
analyzer = StudentAnalyzer(df)

# Create and display comprehensive analysis
fig1 = analyzer.create_comprehensive_analysis()
plt.show()

# Create and display interactive plots
fig2 = analyzer.create_interactive_plots()
plt.show()
```

```python
student_data = df[df["year"]==2023].groupby('student_id')['percent'].first().reset_index().head(10)

print(student_data.shape)
print(student_data)
studentIds = student_data['student_id']
studentPercentages = student_data['percent']

print(f"Unique student IDs: {studentIds.tolist()}")
print(f"Student Percentages: {studentPercentages.tolist()}")

figure, axes = plt.subplots(1, 1, figsize=(30, 20))

axes.plot(studentIds, studentPercentages, color="grey", marker='o', linestyle='-', linewidth=1, markersize=2, label='Student Percentage', markerfacecolor='orange', markeredgewidth=2, markeredgecolor='black', alpha=0.7)
axes.set_xlabel('Student ID')
axes.set_ylabel('Percentage')
axes.set_title('Student Percentages by ID')
axes.legend(loc ='upper right', fontsize=6)
axes.grid(True,color='lightgreen', linestyle='--', linewidth=0.5)

# axes.set_xlim(1,4)
# axes.set_ylim(0,100)
axes.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.subplots_adjust()
plt.show()


figure, axes = plt.subplots(1, 1, figsize=(30, 20))

axes.bar(studentIds, studentPercentages, color="green")
axes.set_xlabel('Student ID')
axes.set_ylabel('Percentage')
axes.set_title('Student Percentages by ID')
axes.legend(loc ='upper right', fontsize=6)
axes.grid(True,color='lightgreen', linestyle='--', linewidth=0.5)

# axes.set_xlim(1,4)
# axes.set_ylim(0,100)
axes.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.subplots_adjust()
plt.show()


figure, axes = plt.subplots(1, 1, figsize=(30, 20))

axes.pie(studentPercentages, labels=studentIds)
axes.set_title('Student Percentages by ID')
axes.legend(loc ='upper right', fontsize=6)
axes.grid(True,color='lightgreen', linestyle='--', linewidth=0.5)
axes.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.subplots_adjust()
plt.show()
```
